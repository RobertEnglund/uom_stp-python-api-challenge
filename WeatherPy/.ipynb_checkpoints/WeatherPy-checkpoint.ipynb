```python
# Dependencies
from citipy import citipy
import json
import matplotlib.pyplot as plt
import pandas as pd
import requests
from config import owm_key
```

```python
# Save config information.
base_url = "http://api.openweathermap.org/data/2.5/weather?"

units = "imperial"

# build partial query URL
query_url = f"{base_url}appid={owm_key}&units={units}&q="
```

```python
# collect data from Open Weather 

# build list of cities to include
cities = ["Algiers", "Antananarivo", "Atjens", "Bamako", "Basseterre",
          "Beijing", "Brasilia", "Brussels", "Cairo", "Canberra", "Cape Town",
          "Gaborone", "Helsinki", "Jakarta", "Kingston",
          "London", "Male", "Manila", "Oslo", "Paris", "Rabat", "Reykjavik",
          "Santiago", "Seoul", "Stockholm", "Tokyo", "Vienna", "Warsaw", "Wellington"
         ]

# set up lists to hold reponse info
valid_cities = []
lat = []
temp = []
humidity = []
wind = []
clouds = []

# Loop through the list of cities and perform a request for data on each
for city in cities:
    response = requests.get(query_url + city).json()
    if response['cod'] == 200:
        valid_cities.append(city)
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        wind.append(response['wind']['speed'])
        clouds.append(response['clouds']['all'])

# build dataframe from results
weather_results_df = pd.DataFrame(
    {"City": valid_cities,
     "Latitude": lat,
     "Temperature": temp,
     "Humidity": humidity,
     "Wind": wind,
     "Clouds": clouds,
     })

weather_results_df
```

```python
# Create scatter plots to show the following relationships from data gathered from Open Weather 
# 1. Temperature (F) vs. Latitude
# 2. Humidity (%) vs. Latitude
# 3. Cloudiness (%) vs. Latitude
# 4. Wind Speed (mph) vs. Latitude
```

```python
# Temperature vs. Latitude
fig = plt.figure()
plt.scatter(temp, lat, marker="o", facecolors="darkblue")
fig.suptitle("Latitude and Temperatures of select cities", fontsize=12)
plt.xlabel('Temperature (F)')
plt.ylabel('Latitude')
plt.show()
```

```python
# Humidity vs. Latitude
fig = plt.figure()
plt.scatter(humidity, lat, marker="o", facecolors="darkblue")
fig.suptitle("Latitude and Humidity of select cities", fontsize=12)
plt.xlabel('Humidity (%)')
plt.ylabel('Latitude')
plt.show()
```

```python
# Cloudiness vs. Latitude
fig = plt.figure()
plt.scatter(clouds, lat, marker="o", facecolors="darkblue")
fig.suptitle("Latitude and Cloudiness of select cities", fontsize=12)
plt.xlabel('Cloudiness (%)')
plt.ylabel('Latitude')
plt.show()
```

```python
# Wind Speed vs. Latitude
fig = plt.figure()
plt.scatter(wind, lat, marker="o", facecolors="darkblue")
fig.suptitle("Latitude and Cloudiness of select cities", fontsize=12)
plt.xlabel('Wind Speed (MPH)')
plt.ylabel('Latitude')
plt.show()
```
